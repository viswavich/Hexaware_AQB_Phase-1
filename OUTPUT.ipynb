```python
from flask import Flask, request, jsonify
from llama_index import SimpleDirectoryReader, LLMPredictor, PromptHelper
import openai
import chromadb
from chromadb.config import Settings

# Initialize the Flask app
app = Flask(__name__)

# Initialize OpenAI API key
openai.api_key = 'your-openai-api-key'

# Initialize Chroma client
chroma_client = chromadb.Client(Settings())

# Create a collection in Chroma (similar to a table in a database)
collection = chroma_client.create_collection(name="documents")

# Load documents from a directory and add embeddings to Chroma
documents = SimpleDirectoryReader('documents').load_data()
for doc in documents:
    embedding = openai.Embedding.create(input=doc.text)['data'][0]['embedding']
    collection.add(doc_id=doc.doc_id, embedding=embedding, metadata={"text": doc.text})

# Helper function to find similar documents based on a query
def search_chroma(query):
    query_embedding = openai.Embedding.create(input=query)['data'][0]['embedding']
    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=5
    )

    # Retrieve the most relevant document texts
    relevant_docs = [result['metadata']['text'] for result in results]
    return relevant_docs

# Route to generate a question based on a topic
@app.route('/generate-question', methods=['POST'])
def generate_question_route():
    data = request.json
    topic = data.get('topic')
    difficulty = data.get('difficulty', 'medium')

    if not topic:
        return jsonify({"error": "Topic is required"}), 400

    # Search for relevant documents using Chroma
    relevant_docs = search_chroma(topic)

    # Generate a question using LlamaIndex
    prompt = f"Based on the following information, create a {difficulty} question:\n\n"
    for doc in relevant_docs:
        prompt += f"{doc}\n\n"

    question = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=150
    ).choices[0].text.strip()

    return jsonify({"question": question})

if __name__ == '__main__':
    app.run(debug=True)
```
